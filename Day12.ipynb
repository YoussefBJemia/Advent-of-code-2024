```python
D = file.split()
R = len(D)
C = len(D[0])

def is_valid(row, col):
    return 0 <= row < R and 0 <= col < C

neighbours = [(-1, 0), (1, 0), (0, -1), (0, 1)]

def calculate_perimeter(row, col):
    """Calculate the perimeter contribution for the given cell."""
    letter = D[row][col]
    cell_perimeter = 4
    for dr, dc in neighbours:
        if is_valid(row + dr, col + dc) and D[row + dr][col + dc] == letter:
            cell_perimeter -= 1
    return cell_perimeter



ans_regions = {}
def create_region(grid):
    regions = []
    visited = set()

    for r in range(R):
        for c in range(C):
            if (r, c) in visited:
                continue
            letter = grid[r][c]
            region = set()
            queue = deque([(r, c)])
            while queue:
                row, col = queue.popleft()
                if not is_valid(row, col) or (row, col) in region or grid[row][col] != letter:
                    continue
                region.add((row, col))
                visited.add((row, col))
                for dr, dc in neighbours:
                    new_row, new_col = row + dr, col + dc
                    if is_valid(new_row, new_col) and grid[new_row][new_col] == letter:
                        queue.append((new_row, new_col))
            regions.append(region)
    return regions


regions = create_region(D)

tot = {}
for i, j in enumerate(regions):
    tot[i] = 0
    area = len(j)
    perimeter = 0
    for r, c in j:
        perimeter += calculate_perimeter(r, c)
    tot[i] = area * perimeter
print(f"result for part 1: {sum(tot.values())}")

# Part 2:
def count_sides(region):
    sides = 0
    for r, c in region:
        # Check each potential side
        for nr, nc, r1, c1, r2, c2 in [
            (r+1, c, r, c-1, r+1, c-1),  # Right side
            (r-1, c, r, c-1, r-1, c-1),  # Left side
            (r, c+1, r-1, c, r-1, c+1),  # Top side
            (r, c-1, r-1, c, r-1, c-1)   # Bottom side
        ]:
            # Count side if neighbor not in region and no diagonal connection
            if (nr, nc) not in region and not ((r1, c1) in region and (r2, c2) not in region):
                sides += 1
    return sides

tot = {}
for i, j in enumerate(regions):
    area = len(j)
    sides = count_sides(j)
    
    tot[i] = area * sides

print(f"result for part 2: {sum(tot.values())}")
```
