```python
import operator

from functools import partial
from itertools import product
from typing import Callable 

operator_mapping = {
    "||": lambda x, y: int(str(x) + str(y)),
    '+': operator.add,
    '*': operator.mul,
}

def left_evaluation(operands: tuple[int], operators: tuple[str]) -> int:
    result = operands[0]
    for op, operand in zip(operators, operands[1:]):
        result = operator_mapping[op](result, operand)
    return result


def parse_data(data: str):
    for line in data.split("\n"):
        result, tail = line.split(": ", maxsplit=1)
        yield int(result), tuple(map(int, tail.split()))


def is_valid(result: int, operands: tuple[int], possible_operations: tuple[str]) -> bool:
    operation_set = list(product(possible_operations, repeat=len(operands) - 1))
    for operations in operation_set:
        evaluation_result = left_evaluation(operands, operations)
        if evaluation_result == result:
            return True


def simulate(data: str, validator: Callable[[tuple, tuple], bool]):
    results = [result for result, operands in parse_data(data) if validator(result, operands)]
    return sum(results)

print("Part 01:", simulate(input, validator=partial(is_valid, possible_operations=("+", "*"))))
print("Part 02:", simulate(input, validator=partial(is_valid, possible_operations=("+", "*", "||"))))
```
