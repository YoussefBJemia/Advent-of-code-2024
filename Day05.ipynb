**<h3>Seperate the input in a list of ordering rules that define the comparison of numbers and a list of updates to be checked for validity<h3>**

```python
def create_rules_update_data(raw_input):
    """
    Parses raw input to extract ordering rules and update lists.
    
    Args:
        raw_input (str): The input string containing rules (e.g., "22|23") 
                         and lists (e.g., "22,23,24").
                         
    Returns:
        dict: A dictionary with 'rules' (list of tuples) and 'lists' (list of lists).
    """
    rules = []
    lists = []
    
    for line in raw_input.strip().split("\n"):
        # If the line contains '|', it belongs to rules
        if '|' in line:
            smaller, bigger = map(int, line.split('|'))
            rules.append((smaller, bigger))
        # If the line contains ',', it belongs to lists
        elif ',' in line:
            lists.append(list(map(int, line.split(','))))
    
    return {"rules": rules, "lists": lists}
```

**<h3>Create class Rule that contains the comparison rule for each number<h3>**

```python
class Rule:
    """
    Defines comparison rules for numbers based on the ordering rules input.
    Generates two lists for each number: numbers smaller and numbers bigger.
    """
    def __init__(self, rules):
        # Expecting rules as a list of tuples [(smaller, bigger), ...]
        self.rules = rules

    def is_smaller_than(self, number):
        """Returns all numbers that are smaller than the given number."""
        return list(set(smaller for smaller, bigger in self.rules if bigger == number))

    def is_bigger_than(self, number):
        """Returns all numbers that are bigger than the given number."""
        return list(set(bigger for smaller, bigger in self.rules if smaller == number))
```

**<h3>Create class Numbers that manages the data<h3>**

```python
class Numbers:
    """
    Represents a collection of numbers and provides operations on them.
    """
    def __init__(self, parsed_data):
        """
        Initializes the Numbers class.
        Args:
            parsed_data (dict): Output from `create_rules_update_data`, with 'rules' and 'lists'.
        """
        self.rules = parsed_data["rules"]  # Rules data
        self.lists = parsed_data["lists"]  # Lists data
        self.unique_numbers = self._extract_unique_numbers()

    def _extract_unique_numbers(self):
        """Extracts all unique numbers from the rules dataset."""
        unique_numbers = set()
        for smaller, bigger in self.rules:
            unique_numbers.add(smaller)
            unique_numbers.add(bigger)
        return sorted(unique_numbers)

    def get_numbers(self):
        """Returns a sorted list of all unique numbers from the rules dataset."""
        return self.unique_numbers

    def get_valid_lists(self, rule_checker):
        """
        Validates the lists based on the given Rule checker.
        Args:
            rule_checker (Rule): An instance of the Rule class to check validity. if the numbers in each row are in the correct order, rule checker == True
        Returns:
            list: Lists from self.lists that are valid.
        """
        valid_lists = []
        for num_list in self.lists:
            is_valid = True
            for i in range(len(num_list) - 1):
                if num_list[i] not in rule_checker.is_smaller_than(num_list[i + 1]):
                    is_valid = False
                    break
            if is_valid:
                valid_lists.append(num_list)
        return valid_lists
    
    def get_invalid_lists(self, rule_checker):
        """
        Validates the lists based on the given Rule checker and returns invalid lists.
        Args:
            rule_checker (Rule): An instance of the Rule class to check validity.
            Returns:
                list: Lists from self.lists that are invalid.
        """
        invalid_lists = []
        for num_list in self.lists:
            is_valid = True
            for i in range(len(num_list) - 1):
                if num_list[i] not in rule_checker.is_smaller_than(num_list[i + 1]):
                    is_valid = False
                    break
            if not is_valid:
                invalid_lists.append(num_list)
        return invalid_lists   
```

**<h3>Define a function to sum the middle values of each valid list<h3>**

```python
def sum_middle_values(valid_lists):
    """
    Sums the values in the middle position (1-based index: n+1/2) of each list in valid_lists.
    
    Args:
        valid_lists (list): A list containing lists of numbers.
        
    Returns:
        int: The total sum of the middle values.
    """
    total = 0
    for value in valid_lists:
        n = len(value)
        if n % 2 == 1:  # Odd-length list
            middle_value = value[n // 2]  # 0-based index for middle
            total += middle_value
        else:  # Even-length list
            raise ValueError("The lists must have an odd number of elements to find a single middle value.")
    return total
```

**<h3>Bubble sort unvalidated lists<h3>**

```python
def order_invalid_lists(invalid_lists, rule_checker):
    """
    Orders the invalid lists by applying a bubble sort based on the is_smaller_than rule.
    Args:
        invalid_lists (list): The list of invalid lists.
        rule_checker (Rule): An instance of the Rule class to check validity.
    Returns:
        list: A list of sorted invalid lists.
    """
    sorted_invalid_lists = []  # To store the sorted lists
    
    for num_list in invalid_lists:
        n = len(num_list)
        sorted_list = num_list[:]  # Make a copy of the list to sort
        
        # Bubble sort logic
        for i in range(n - 1):
            for j in range(n - 1 - i):  # Reduce the range with each pass
                if sorted_list[j] not in rule_checker.is_smaller_than(sorted_list[j + 1]):
                    # Swap the elements if they are not in correct order
                    sorted_list[j], sorted_list[j + 1] = sorted_list[j + 1], sorted_list[j]
        
        # Add the sorted list to the result
        sorted_invalid_lists.append(sorted_list)
    
    return sorted_invalid_lists
```
