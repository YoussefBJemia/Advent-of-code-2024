```python
import numpy as np

# Convert the input grid into a NumPy array for easier manipulation
rows = input_grid.split('\n')
matrix = np.array([list(row) for row in rows])

print("Initial Grid:")
print(matrix)
```

```python
# Class representing the Guard and its movement logic
class Guard:
    def __init__(self, x, y, direction):
        """
        Initialize the guard's position and direction.
        Args:
            x (int): Initial x-coordinate.
            y (int): Initial y-coordinate.
            direction (str): Initial direction ('^', 'v', '<', '>').
        """
        self.x = x
        self.y = y
        self.direction = direction

    def direction_to_vector(self):
        """
        Map the current direction to a movement vector.
        Returns:
            tuple: Movement vector (dx, dy).
        """
        if self.direction == '^':
            return (0, -1)
        elif self.direction == 'v':
            return (0, 1)
        elif self.direction == '<':
            return (-1, 0)
        elif self.direction == '>':
            return (1, 0)
        else:
            raise ValueError('Invalid direction')

    def move(self):
        """
        Move the guard one step in the current direction.
        """
        vector = self.direction_to_vector()
        self.x += vector[0]
        self.y += vector[1]

    def move_back(self):
        """
        Move the guard one step back in the opposite direction.
        """
        vector = self.direction_to_vector()
        self.x -= vector[0]
        self.y -= vector[1]

    def turn(self):
        """
        Turn the guard 90 degrees clockwise.
        """
        if self.direction == '^':
            self.direction = '>'
        elif self.direction == 'v':
            self.direction = '<'
        elif self.direction == '<':
            self.direction = '^'
        elif self.direction == '>':
            self.direction = 'v'
        else:
            raise ValueError('Invalid direction')
```

```python
#Initial position of the guard 
initial_x = None
initial_y = None

for row_index in range(len(matrix)):
    for col_index in range(len(matrix[row_index])):
        if matrix[row_index][col_index] not in {'#', '.'}:
            initial_y = row_index
            initial_x = col_index
            break
    if initial_y is not None:
        break

print("Initial position of the guard (x, y):", (initial_x, initial_y))

initial_direction = matrix[initial_y][initial_x]
print("Initial direction of the guard:", initial_direction)

# Determine the grid's vertical and horizontal borders
vertical_borders = (0, len(matrix[0]) - 1)  # Min and max column indices
horizontal_borders = (0, len(matrix) - 1)  # Min and max row indices
print("Vertical Borders (min_x, max_x):", vertical_borders)
print("Horizontal Borders (min_y, max_y):", horizontal_borders)
```

**<h3>Part I: Distinct positions<h3>**

```python
# Initialize the Guard object with its position and direction
guard = Guard(initial_x, initial_y, initial_direction)
distinct_positions = set()

while True:
    guard.move() # Move the guard one step in the current direction
    #print(f"Guard Position: ({guard.x}, {guard.y}), Cell: {matrix[guard.y][guard.x]}, Direction: {guard.direction}")

    if matrix[guard.y][guard.x] == "#": # If the guard encounters an obstruction '#', move back and turn
        guard.move_back()
        guard.turn()
    else:
        distinct_positions.add((guard.x, guard.y)) # Add the current position to the set of distinct positions

    if guard.x in horizontal_borders or guard.y in vertical_borders: # Check if the guard has exited the map by checking whether his last position was on the border
        print(f"The guard exited the map after visiting {len(distinct_positions)} distinct positions.")
        break
```

**<h3>Part II: obstructions<h3>**

<p> <b> Lazy method (grid search)</b>: Set up a grid the same size as the matrix to track the positions the guard visits, except the starting point.</p>
<p>Try placing an obstruction in each position and check if the guard can reach the border of the map.</p>
<p>If the guard can reach the border, the obstruction doesn’t matter. If not, it’s considered useful and the obstruction is added to non_stopping_grids.</p>
<p>Since infinite loops are hard to catch, we use a quick workaround: set a max number of steps or running time. If the guard doesn’t stop within that limit, it’s probably stuck in an infinite loop.</p>

```python
# Create a list of positions where obstructions can be placed, excluding the guard's initial position
obstruction_grid = distinct_positions.copy()
obstruction_grid.remove((initial_x, initial_y))
```

```python
import time

def find_non_stopping_obstructions(max_steps):
    """
    Identifies grid positions that cause the guard simulation to exceed max_steps.

    Args:
        max_steps (int): Maximum allowed steps in the simulation.

    Returns:
        None: Prints the number of non-stopping obstructions and the runtime.
    """
    start_time = time.time()  # Record the start time for performance tracking
    
    # List to store configurations that cause the program not to stop
    non_stopping_grids = []

    # Iterate through each position in the obstruction grid
    for obstruction_x, obstruction_y in obstruction_grid:
        # Backup the original value at the current position
        original_value = matrix[obstruction_y][obstruction_x]
        
        # Place an obstruction ('#') at the current grid position
        matrix[obstruction_y][obstruction_x] = '#'

        # Initialize the guard with its starting position and direction
        guard = Guard(initial_x, initial_y, initial_direction)
        visited_positions = set()  # Track visited positions
        step_count = 0  # Counter for the number of steps taken
        stopped = False  # Flag to determine if the program stopped
        
        # Simulate guard movements
        while step_count <= max_steps:
            guard.move()
            
            if matrix[guard.y][guard.x] == "#":
                guard.move_back()
                guard.turn()
            else:
                visited_positions.add((guard.x, guard.y))
            
            # Check if the guard exits the map
            if guard.x in horizontal_borders or guard.y in vertical_borders:
                stopped = True
                break
            
            step_count += 1  # Increment the step count
        
        # Check if the guard did not stop within the allowed steps
        if step_count > max_steps:
            non_stopping_grids.append((obstruction_x, obstruction_y))
        
        # Restore the original grid value
        matrix[obstruction_y][obstruction_x] = original_value

    # Calculate runtime
    running_time = time.time() - start_time

    
    print(f"For a maximum of {max_steps} steps within each loop, the program recorded {len(non_stopping_grids)} possible non-stopping obstructions.")
    print(f"The running time was {running_time:.2f} seconds.")
```

```python
find_non_stopping_obstructions(1000)
find_non_stopping_obstructions(5000)
find_non_stopping_obstructions(10000)
find_non_stopping_obstructions(20000)
```
