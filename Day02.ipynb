```python
import re
def nums(s):
    return [int(match.group(0)) for match in re.finditer(r'-?\d+', s)]
```

```python
#part 1:
R = file.split("\n")

def is_safe(report):
    safe = all(i > j for i, j in zip(report, report[1:])) or all(i < j for i, j in zip(report, report[1:]))
    if safe:
        report = sorted(report)
        for i, j in zip(report, report[1:]):
            if j-i > 3 or j-i < 1:
                safe = False
                break
    return safe

ans = 0
for line in R:
    report = nums(line)
    safe = is_safe(report)
    if safe:
        ans += 1    

print(ans)
```
