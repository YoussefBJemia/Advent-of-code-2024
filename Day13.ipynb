```python
import re
import math
import numpy as np
import time
def nums(s):
  return [int(match.group(0)) for match in re.finditer(r'\d+', s)]
```

```python
def no_solution(a, b, p):
  for i in range(len(p)):
    gcd_value = math.gcd(a[i], b[i])
    if p[i] % gcd_value != 0:
      return True
  return False

cost_a = 3
cost_b = 1
tokens = 0

start = time.time()
for l in file.split("\n\n"):
  a, b, p = l.split("\n")
  a = nums(a)
  b = nums(b)
  p = nums(p)
  if no_solution(a, b, p):
    #print("no solution")
    continue
  #print(a, b, p)
  best = 10000
  for pressa in range(101):
    for pressb in range(101):
      #print(f"{[a, b]}")
      lhs = np.dot(pressa, a) + np.dot(pressb, b)
      if np.all(lhs == p):
        best = min(cost_a * pressa + cost_b * pressb, best)
        #print(f"price is {price}")
  if best < 10000:
    tokens += best
print(f"time: {time.time() - start}")
print(tokens)
```
